```python
#Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above
corr = sns.heatmap(df_sub.corr(), annot = True) 
```

###  With :Average age of mother: A 0.44 coefficient indicates a medium-to-strong positive correlation. Indicates   Average Birth has an upward trend with Average age of Mother.
### With :Education Level code: A 0.52 coefficient indicates a medium-to- strong positive correlation. This also indicates Average Birth has an upward trend with Education level code. 
### With: Number of births: A 0.041 coefficient indicates no correlation between number of births and Average Birth weight.

# 06 Create a categorical plot and interpret the results.

```python
#Creating a categorical variable that splits the \"Avearge Birth weight\" column into categories
sns.histplot(df_sub['Average Birth Weight (g)'], bins = 20, kde = True)
```

```python
df_sub.loc[df_sub['Average Birth Weight (g)'] < 2600, 'Average weight category'] = 'Low weight'
```

```python
df_sub.loc[(df_sub['Average Birth Weight (g)'] >= 2600) & (df_sub['Average Birth Weight (g)'] < 3000), 'Average weight category'] = 'medium weight'
```

```python
df_sub.loc[df_sub['Average Birth Weight (g)'] >= 3000, 'Average weight category'] = 'High weight'
```

```python
df_sub['Average weight category'].value_counts(dropna = False)
```
